## 최솟값 구하기 

```python
arr = [5, 3, 7, 9, 2, 5, 2, 6]
```

```python
# solution 1 

arrMin = float('inf') # 가장 큰 값이 저장될 수 있도록 초기화

for i in range(len(arr)):
    if arr[i] < arrMin:
        arrMin = arr[i]
    else:
        pass

print(arrMin)
```

```python
# solution 2 

arrMin = float('inf') # 가장 큰 값이 저장될 수 있도록 초기화

for x in arr:
    if x < arrMin:
        arrMin = x 

print(arrMin)
```

## 정다면체 

두 개의 정 N면체와 정 M면체의 두 개의 주사위를 던져서 나올 수 있는 눈의 합 중 가장 확률이 높은 숫자를 출력하는 프로그램을 작성하세요. 
정답이 여러 개일 경우 오름차순으로 출력합니다.

- 입력 설명:
첫 번째 줄에는 자연수 N과 M이 주어집니다. N과 M은 4,6,8,12,20중의 하나입니다.
- 출력 설명 :
첫 번째 줄에 답을 출력합니다

```python
n = 4 
m = 6 
```

```python
cnt = [0]*(n + m +1 )
max = 0

for i in range(1, n+1):
    for j in range(1, m+1):
        cnt[i+j] += 1 

for i in range(n + m + 1): # finding max_cnt
    if cnt[i] > max:
        max = cnt[i]

for i in range(n + m + 1): # finding value
    if cnt[i] == max:
        print(i, end = '') 
```
