## [문제116] 근무일수가 가장 많은 10위까지 직원들의 employee_id, last_name, department_name, 근무일수를 출력해주세요.

```python
# 라이브러리 불러오기 
import pandas as pd 
import datetime 
from pandas import Series, DataFrame
```

```python
# 데이터 불러오기 
dept = pd.read_csv("c:/data/departments.csv")

# date 날짜 format 맞춰서 불러오기 
emp = pd.read_csv("c:/data/employees.csv", parse_dates = ['HIRE_DATE'], # 옵션 설정 
                 date_parser = lambda arg: pd.to_datetime(arg, format = '%Y-%m-%d'))
```

```python
# 이름과 근무일수 뽑아주기 
x = DataFrame({'사번' : emp['EMPLOYEE_ID'],
                '이름':emp['LAST_NAME'],
                  '부서코드' : emp['DEPARTMENT_ID'],                 
               '근무일수':(pd.Timestamp.now() - emp['HIRE_DATE']).dt.days})
```

```python
# 순위 컬럼 추가하기

x['순위'] = x['근무일수'].rank(ascending = False, method = 'dense').astype('int')
```

```python
# rank 10위까지 출력하기 
top10 = x.loc[x['순위'] <= 10,].sort_values(by='순위', ascending = True)
top10
```

```python
# 부서 이름 함께 출력하기
result = pd.merge(top10, dept, left_on = '부서코드', right_on = 'DEPARTMENT_ID')[['사번','이름','부서코드','DEPARTMENT_NAME','근무일수','순위']]
result = result.rename(columns = {'DEPARTMENT_NAME' : '부서이름'})
result.sort_values(by='순위')
```

## [118번] 관리자가 아닌 사원의 정보를 출력해주세요. 

```python
# 서브쿼리 
c.execute('''select *
            from emp
            where employee_id not in (select distinct manager_id 
                                     from emp
                                     where manager_id is not null) # null을 제어하지 않으면 결과가 출력되지 않음 
            ''')
c.fetchall()
```

## [문제120] 부서이름별 총액급여를 출력해주세요

```python
c.execute('''SELECT department_name,
                    sum(salary)
            FROM dept 
                RIGHT JOIN emp ON dept.department_id = emp.department_id
            GROUP BY department_name
            ORDER BY 2 DESC''')
c.fetchall()
```

## [문제 121] 자신의 부서 평균 급여보다 더 많이 받는 사원들의 정보를 출력해주세요 

```python
c.execute('''select e2.*
            from (select department_id, avg(salary) avgsal
                  from emp
                  group by department_id) e1, emp e2
            where e1.department_id = e2.department_id
            and e2.salary > e1.avgsal''')
c.fetchall()
```

## [문제 122] 근무 일수가 가장 많은 10위까지 직원들의 employee_id, last_name, department_name, 근무 일수를 출력하시오

```python
c.execute(''' 
            select employee_id, last_name, department_name, workingday
            from (
                    select employee_id, last_name, department_id, rnk, workingday
                    from (
                        select employee_id, last_name, department_id,
                                dense_rank() over(order by workingday desc) rnk,
                                workingday
                        from (
                                select employee_id, last_name, department_id,
                                        (strftime('%s','now') - strftime('%s',date(hire_date)))/86400 workingday
                                from emp)
                    )
                    where rnk <= 10 ) e, dept d 
            where e.department_id = d.department_id 
            order by 4 desc
            ''')
c.fetchall()
```

## sqlite 
**장점 : db가 필요가 없고, 응용프로그램 안에서 사용 가능한 가벼운 db**
- 별도의 DB서버가 필요없이 DB파일을 기초로 하여 데이터베이스를 처리하는 엔진
- 응용프로그램에서 가볍게 사용되는 데이터베이스이다.
- 현장에서 써봤냐 자주 묻는다

```python
import sqlite3
```

```python
# db를 구성하기 위한 접속! 
conn = sqlite3.connect(':memory:') #  dbms를 메모리에 만들고 버리겠다 = python 프로그램 쓰는 동안만 만들고 버리겠다
conn 
```

```python
# transaction : 논리적으로 DML(insert, update, delete, merge)문을 한꺼번에 처리하는 작업단위 
# transaction 종료 : commit, rollback 

conn.rollback() # 커서가 아닌, coonect를 기반으로 한 rollback을 해야한다. 
```

```python
# sqllite을 종료할 때는 cursor -> conn 종료해야 함. 
## 위에서 insert한 데이터들은 memory가 종료되어 사라지게 됨 
c.close()
conn.close()
```
